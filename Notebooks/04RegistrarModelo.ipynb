```python
catalog = "main"
schema = "dbdemos_fs_travel"

# Experimento y modelo
xp_name = f"compra_viaje_experimento_{datetime.now().strftime('%Y-%m-%d_%H:%M:%S')}"
xp_path = "/Workspace/Proyectos_Dev/databricks-medallion/Experiments"
model_name = "purchase_travel_model"
model_full_name = f"{catalog}.{schema}.{model_name}"
```

```python
import mlflow
from mlflow import MlflowClient

mlflow.set_registry_uri("databricks-uc")
client = MlflowClient()

run_id = mlflow.search_runs(filter_string=f"tags.mlflow.project.name = '{xp_name}'", max_results=1)["run_id"].values[0]
latest_model = mlflow.register_model(f"runs:/{run_id}/model", f"{model_full_name}")
client.set_registered_model_alias(name=f"{model_full_name}", alias="Champion", version=latest_model.version)
```
