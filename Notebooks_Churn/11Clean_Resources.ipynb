# Limpiar recursos como experimentos y modelos

```python
catalog = "medallion_dev"
schema = "gold"
model_name = "customer_churn_classifier"
project_path = "/Workspace/Proyectos_Dev/databricks-medallion"
model_full_name = f"{catalog}.{schema}.{model_name}"

xp_name = f"{model_name}"
xp_path = f"{project_path}/Experiments"
experiment_full_path = f"{xp_path}/{xp_name}"
```

```python
from mlflow import MlflowClient

client = MlflowClient()

delete_model = True
try:
    model_versions = client.search_model_versions(f"name='{model_full_name}'")
    if not model_versions:
        print(f"✅ Modelo: {model_full_name} no tiene versiones")
    else:
        for version in model_versions:
            client.delete_model_version(model_full_name, version.version)
        print(f"✅ Todas las versiones del modelo: {model_full_name} borradas")
    if delete_model:
        client.delete_registered_model(model_full_name)
        print(f"✅ Modelo: {model_full_name} borrado")
    else:
        print(f"✅ Modelo: {model_full_name} NO borrado porque la bandera de borrado = False")
except Exception as e:
    print(f"Error al borrar modelo: {model_full_name}. Mensaje: {e}")
```

```python
import mlflow

delete_experiment = True
try:
    experiment = mlflow.get_experiment_by_name(experiment_full_path)    
    if experiment is None:
        print(f"✅ Experimento: {experiment_full_path} no existe")
    else:
        if delete_experiment:
            mlflow.delete_experiment(experiment.experiment_id)
            print(f"✅ Experimento: {experiment_full_path} borrado")
        else:
            print(f"✅ Experimento: {experiment_full_path} NO borrado porque la bandera de borrado = False")
except Exception as e:
    print(f"Error al borrar experimento: {experiment_full_path}. Mensaje: {e}")
```
