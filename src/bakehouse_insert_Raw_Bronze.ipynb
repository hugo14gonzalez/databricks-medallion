###Borrar y crear widgets

```python
dbutils.widgets.removeAll()
dbutils.widgets.text("catalog", "bakehouse_dev")

catalogName = dbutils.widgets.get("catalog")
```

### Insertar - Raw

```python
print(f"Insertando datos en: {catalogName}.raw.customers_raw...")
spark.sql(f"""
INSERT INTO {catalogName}.raw.customers_raw (customerID, first_name, last_name, email_address, phone_number, address, city, state, country, continent, postal_zip_code, gender) 
SELECT customerID, first_name, last_name, email_address, phone_number, address, city, state, country, continent, postal_zip_code, gender
FROM samples.bakehouse.sales_customers
""")

print(f"Insertando datos en: {catalogName}.raw.franchises_raw...")
spark.sql(f"""
INSERT INTO {catalogName}.raw.franchises_raw (franchiseID, name, city, district, zipcode, country, size, longitude, latitude, supplierID)
SELECT franchiseID, name, city, district, zipcode, country, size, longitude, latitude, supplierID
FROM samples.bakehouse.sales_franchises
""")

print(f"Insertando datos en: {catalogName}.raw.suppliers_raw...")
spark.sql(f"""
INSERT INTO {catalogName}.raw.suppliers_raw (supplierID, name, ingredient, continent, city, district, size, longitude, latitude, approved)
SELECT supplierID, name, ingredient, continent, city, district, size, longitude, latitude, approved
FROM samples.bakehouse.sales_suppliers
""")

print(f"Insertando datos en: {catalogName}.raw.transactions_raw...")
spark.sql(f"""
INSERT INTO {catalogName}.raw.transactions_raw (transactionID, customerID, franchiseID, dateTime, product, quantity, unitPrice, totalPrice, paymentMethod, cardNumber)
SELECT transactionID, customerID, franchiseID, dateTime, product, quantity, unitPrice, totalPrice, paymentMethod, cardNumber
FROM samples.bakehouse.sales_transactions
""")
```
