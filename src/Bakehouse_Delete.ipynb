### Borrar y crear widgets

```python
dbutils.widgets.removeAll()
dbutils.widgets.text("catalog", "bakehouse_dev")

catalogName = dbutils.widgets.get("catalog")
```

### Borrar tablas origenes (raw)

```python
try:
    if any(catalog.name == catalogName for catalog in spark.catalog.listCatalogs()):
        spark.catalog.setCurrentCatalog(catalogName)
        print(f"Catalgo actual: {spark.catalog.currentCatalog()}")

        schemaNames = ["raw", "bronze", "silver", "gold"]
        for schemaName in schemaNames:            
            print(f"Borrando tablas para el esquema: {schemaName}...")
            if spark.catalog.databaseExists(schemaName):
                for table in spark.catalog.listTables(schemaName):
                    spark.sql(f"DROP TABLE IF EXISTS {catalogName}.{schemaName}.{table.name}")
    else:
        print(f"El catalogo {catalogName} no existe.")
except Exception as e:
    msg = f"Error al borrar tablas en el catalog: {catalogName}. Mensaje: {e}"
    #print(msg)
    raise Exception(msg)
```
